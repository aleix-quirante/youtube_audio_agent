# Imports

```python
import whisper
import yt_dlp
import os
```

# Audio download

```python
#Audio folder
os.makedirs("../data/raw_audio", exist_ok=True)

#Url
video_url = "https://www.youtube.com/watch?v=YWIe4Sqr0Ts" 

#Select only audio and convert to mp3
ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '320',
    }],
    #Save the file with the title of the video and .mp3 extension in the raw_audio folder
    'outtmpl': '../data/raw_audio/%(title)s.%(ext)s',
    'quiet': False #Progress bar and other info will be printed
}

print(f"Start audio download: {video_url}")

#Download and convert the audio
try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([video_url])
    print("✅ Audio downloaded and converted successfully!")
except Exception as e:
    print(f"❌ Error occurred: {e}")
```

# Voice Transcription

```python
audio_path = "../data/raw_audio/Breaking Down Eminem & Juice WRLD’s “Godzilla” Video ｜ Genius News.mp3"

print("Loading Whisper")
#Model Whisper
model = whisper.load_model("base") 

print(f"Transcribing!")

#Transcription 
try:
    result = model.transcribe(audio_path)
    texto_trans = result["text"]
    
    print("\n✅ ¡Tranascrption finished successfully! Here's a preview of the transcribed text:\n")
    print(texto_trans[:500] + "...\n")
    
    #Save in text file
    os.makedirs("../data/transcripts", exist_ok=True)
    with open("../data/transcripts/godzilla_transcript.txt", "w", encoding="utf-8") as f:
        f.write(texto_trans)
    print("The complete text has been saved in: ../data/transcripts/godzilla_transcript.txt")

except Exception as e:
    print(f"❌ Error: {e}")
```
